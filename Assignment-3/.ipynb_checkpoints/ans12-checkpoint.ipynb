```python
import pandas as pd
```

```python
# i) Read the dataset in dataframe.
df = pd.read_csv('./Salary_Data.csv')

# ii) Display first five records.
print("First five records:")
print(df.head())

# iii) Display first ten records.
print("\nFirst ten records:")
print(df.head(10))

# iv) Display last five records.
print("\nLast five records:")
print(df.tail())

# v) Display last ten records.
print("\nLast ten records:")
print(df.tail(10))

# vi) Display the columns inside the dataset.
print("\nColumns in the dataset:")
print(df.columns)

# vii) Display shape of data.
print("\nShape of the dataset:")
print(df.shape)

# viii) Describe the dataset.
print("\nDescription of the dataset:")
print(df.describe())

# ix) Display the information about the dataset and analyze the data.
print("\nInformation about the dataset:")
print(df.info())

# x) Display types of each column.
print("\nData types of each column:")
print(df.dtypes)

# xi) Find out maximum, minimum, mean, median, standard deviation value of each column.
print("\nStatistical values of each column:")
stats = df.agg(['max', 'min', 'mean', 'median', 'std'])
print(stats)
```
